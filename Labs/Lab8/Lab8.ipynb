# Part A

Create a function that returns the SFR for a given luminosity (NUV, FUV, TIR, Halpha)

$Log( {\rm SFR} (M_\odot/year)) = Log(Lx (erg/s)) - Log(Cx)$ 

Including corrections for dust absorption 

Kennicutt & Evans 2012 ARA&A Equation 12 and Table 1, 2

```python
def StarFormationRate(L, Type, TIR=0):
    '''
    Compute the star fomration rate (SFR) of a galaxy following Kennicutt & Evans 2012, Eq. 12 (ARA&A 50)
    Parameters:
        Inputs:
            L: float
                Luminosity of a galaxy in erg/s
            Type: string
                Wavelenth options:
                    'FUV', 'NUV', 'TIR', 'Halpha'
            TIR: float
                Total infrared luminosity in erg/s (default = 0)
        Output:
            SFR: float
                Log of the star formation rate (Msun/yr)
    '''
    # Make corrections for constants and dust absorption
    if (Type == 'FUV'):
        # Calibration from LFUV to SFR
        logCx = 43.35 
        # Correction for dust absorption
        TIRc = 0.46
    elif (Type == 'NUV'):
        logCx = 43.17
        TIRc = 0.27
    elif (Type == 'Halpha'):
        logCx = 41.27
        TIRc = 0.0024
    elif (Type == 'TIR'):
        logCx = 43.41
        TIRc = 0.0
    else:
        print("Error: Wavelength Type must be 'FUV', 'NUV', 'TIR', or 'Halpha'")
        sys.exit()

    # Correct the luminosity for dust using the TIR
    Lcorr = L + TIRc*TIR

    # Star formation rate
    SFR = np.log10(Lcorr) - logCx

    return SFR
```

# Part B Star formation main sequence

1) Write a function that returns the average SFR of a galaxy at a given redshift, given its stellar mass

2) What is the average SFR of a MW mass galaxy today? at z=1?

3) Plot the SFR main sequence for a few different redshifts from 1e9 to 1e12 Msun.


From Whitaker 2012:

log(SFR) = $\alpha(z)({\rm log}M_\ast - 10.5) + \beta(z)$

$\alpha(z) = 0.7 - 0.13z$

$\beta(z) = 0.38 + 1.14z - 0.19z^2$

# Step 1

```python
def SFRMainSequence(Mstar, z):
    '''
    Compute the average SFR of a galaxyas a function of stellar mass and redshift
    Inputs:
        Mstar: float
            Stellar mass of the glaxy in units of Msun
        z: float
            redshift
    Output:
        SFR: float
            Log of the SFR (Msun/yr)
    '''
    alpha = 0.7 - 0.13*z
    beta = 0.38 + 1.14*z - 0.19*z**2

    SFR = alpha * (np.log10(Mstar) - 10.5) + beta

    return SFR
```

# Step 2

```python
# MW stellar mass at z=0
MWmstar = 7.5e10

# SFR for MW type galaxy at z=0
print(SFRMainSequence(MWmstar, 0))
print(10**SFRMainSequence(MWmstar, 0))
```

```python
# SFR for MW type galaxy at z=1
print(SFRMainSequence(MWmstar, 1))
print(10**SFRMainSequence(MWmstar, 1))
```

```python
# create an array of stellar masses
mass = np.linspace(1e8, 1e12)
```

```python

fig = plt.figure(figsize=(8,8), dpi=500)
ax = plt.subplot(111)

# add log log plots
# z=0
plt.plot(np.log10(mass), SFRMainSequence(mass, 0), color='b', label='z=0')
# z=1
plt.plot(np.log10(mass), SFRMainSequence(mass, 1), color='r', label='z=1')
# z=2
plt.plot(np.log10(mass), SFRMainSequence(mass, 2), color='g', label='z=2')
# z=3
plt.plot(np.log10(mass), SFRMainSequence(mass, 3), color='magenta', label='z=3')

# Add axis labels
plt.xlabel('Log(Mstar (M$_\odot$))', fontsize=12)
plt.ylabel('Log(SFR (M$_\odot$/year))', fontsize=12)


#adjust tick label font size
label_size = 12
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size

# add a legend with some customizations.
legend = ax.legend(loc='upper left',fontsize='x-large')


# Save file
plt.savefig('Lab8_SFR_MainSequence.png')
plt.show()
```
